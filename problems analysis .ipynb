```python
def val (vd):
    num=vd
    if (num <=5).bool():
        return 'low'
    elif (num >5).bool() and (num <=7).bool() :
        return 'medium'
    else :
        return 'hight'
```

```python
df['Value']=df[['Value added (1-10)']].apply(val,axis=1)
```
